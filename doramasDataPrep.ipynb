```python
import json

# Dados da lista (suponho que você já tenha definido a lista_informacoes)

# Função para extrair informações de cada item
def extrair_informacoes(item):
    item_info = {}
    lines = item.strip().split('\n\n')
    for line in lines:
        parts = line.split(': ', 1)
        if len(parts) == 2:
            key, value = parts
            item_info[key] = value
    return item_info

# Converter para JSON
lista_json = [extrair_informacoes(item) for item in lista_informacoes]

json_data = json.dumps(lista_json, indent=4, ensure_ascii=False)

# Escrever o JSON em um arquivo
with open('resultado_de_conversao_doramas.json', 'w', encoding='utf-8') as f:
    f.write(json_data)
```
